```python
import pandas as pd
df=pd.read_csv('spam.csv')
df
```

```python
df["Message"][2]
```

```python
text=df['Message'][5]
```

```python
text
```

```python
import re
```

```python
# Remove special characters and punctuation using regex
cleaned_text = re.sub(r'[^\w\s]', '', text)  # Removes everything except word characters and spaces
cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Replaces multiple spaces with a single space
cleaned_data= cleaned_text.strip()  # Removes leading/trailing whitespace
```

```python
cleaned_data
```

```python
mydata=[]
for i in df['Message']:
  cleaned_text = re.sub(r'[^\w\s]', '', i)  # Removes everything except word characters and spaces
  cleaned_text = re.sub(r'\s+', ' ', cleaned_text)  # Replaces multiple spaces with a single space
  cleaned_data= cleaned_text.strip()
  mydata.append(cleaned_data)
```

```python
mydata
```

```python
from nltk import word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
tokens=[word_tokenize(i) for i in mydata]
```

```python
tokens
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop=stopwords.words('english')
```

```python
tokens[0]
```

```python
[i for i in tokens[0] if i not in stop]
```

```python
#x is message
data=[]
for x in tokens:
  cl=[i for i in x if i not in stop]
  data.append(cl)
```

```python
data
```

```python
#lemmatization is simplifying using the single base or root form of verb forms with a meaning
#stemming is removing the prefix and suffix of the words with the meaning
```

```python
from nltk.stem import PorterStemmer
```

```python
x=data[0][4]
```

```python
stem=PorterStemmer()
```

```python
x
```

```python
stem.stem(x)
```

```python
stemmed_data=[]
for i in data:
  dr=[stem.stem(x) for x in i]
  stemmed_data.append(dr)
```
